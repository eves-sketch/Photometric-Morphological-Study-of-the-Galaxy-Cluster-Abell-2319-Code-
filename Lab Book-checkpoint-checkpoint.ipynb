DARK SUBTRACTION

```python
#checking all exposure times to see if i have to scale any

for f in flatRlist2319:
    flatr = CCDData.read(f, unit = 'adu')
    print(f"{f}: EXPTIME = {flatr.header['EXPTIME']} s") # will be the same for b

for fb in flatBlist2319:
    flatb = CCDData.read(fb, unit = 'adu')
    print(f"{fb}: EXPTIME = {flatb.header['EXPTIME']} s")
    
for b in darklist2319:
    dark = CCDData.read(b, unit = 'adu')
    print(f"{b}: EXPTIME = {dark.header['EXPTIME']} S")
```

```python
# NOW SUBTRACTING THE MASTER DARK FRAME FROM THE FLAT FRAMES IN THE R AND B BANDS AND THEN NORMALISING THEM


fbdsubB = [subtract_dark(b, darkbias2319frame, exposure_time = 'EXPTIME', exposure_unit = u.second, scale = True) for b in flatbiassubB]
fbdsubR = [subtract_dark(r, darkbias2319frame, exposure_time = 'EXPTIME', exposure_unit = u.second, scale = True) for r in flatbiassubB]

finalflatb = ccdproc.combine(fbdsubB, method='median')
finalflatr = ccdproc.combine(fbdsubR, method='median')
#------------------------------------------------------------------------------------------------------------------------------------------

meancountsflatB = np.median(finalflatb, axis = 0)
normflatB = finalflatb / (meancountsflatB)
norm_flatB = CCDData(normflatB, unit='adu')

meancountsflatR = np.median(finalflatr, axis = 0)
normflatR = finalflatr / (meancountsflatR)
norm_flatR = CCDData(normflatR, unit='adu')

```

BIAS AND DARK SUBTRACTION FROM SCIENCE FRAME AND APPENDING TO A LIST

```python
#CALLING IN SCIENCE FRAMES FOR ABELL 2319 AND APPENDING TO A LIST

scienceBpath = r"C:\Users\exs177\OneDrive - University of Birmingham\CompAstro2025\science_B\*.fit"
scienceBlist = glob.glob(scienceBpath) #combines all science.fit files into one list ready to be iterated through the loop below

scienceRpath = r"C:\Users\exs177\OneDrive - University of Birmingham\CompAstro2025\science_R\*.fit"
scienceRlist = glob.glob(scienceRpath)

#----------------------------------------------------------
#CREATING THE CALIBRATED BACKGROUND FRAMES FOR B SCIENCE AND R SCIENCE FRAMES

calibrated_framesB = []

for sBfile in scienceBlist:
    sciB = CCDData.read(sBfile, unit='adu') #reads all science files in the list
    
    # Subtract bias
    bias_subB = sciB.subtract(mastbi2319Frame) #subtracts the master bias iteratively
    #bias_subB = subtract_bias(sciB, mastbi2319Frame)
    
    # Scale dark to science exposure
    dark_exposure = darkbias2319frame.header['EXPTIME'] * u.second
    t_sciB = sciB.header['EXPTIME'] * u.second
    scaled_darkB = darkbias2319frame.multiply(t_sciB / dark_exposure)
    
    # Subtract scaled dark
    dark_subB = bias_subB.subtract(scaled_darkB) #subtracting scaled master dark
    
    # Divide by normalized flat
    flat_correctedB = dark_subB.divide(norm_flatB) #dividing by normalised flat
    
    calibrated_framesB.append(flat_correctedB) #appeneding back into a list called "calibrated frames"

#we repeat this for the r filtewr science frames
#----------------------------------------------------------------------------

calibrated_framesR = []

for sRfile in scienceRlist:
    sciR = CCDData.read(sRfile, unit='adu')
    
    # Subtract bias
    bias_subR = subtract_bias(sciR, mastbi2319Frame)
    
    # Scale dark to science exposure
    dark_exposure = darkbias2319frame.header['EXPTIME'] * u.second
    t_sciR = sciR.header['EXPTIME'] * u.second
    scaled_darkR = darkbias2319frame.multiply(t_sciR / dark_exposure)
    
    # Subtract scaled dark
    dark_subR = bias_subR.subtract(scaled_darkR)
    
    # Divide by normalized flat
    flat_correctedR = dark_subR.divide(normflatR)
    
    calibrated_framesR.append(flat_correctedR)
```

SOURCE EXTRACTION  FOR THE B-FILTER

```python
#GETTING IMAGE DIMENSIONS

dataB = frame_135B.data
# Get the shape (dimensions)
print("Image dimensions:", dataB.shape)
```

This is so we know what box size we would need when we create the background Image. It has to be a multiple of 4096 and not too small or big. If the box is too small, the resulting background map will be too noisy with artificial fluctuations. If the box size is too big we might smooth over real background variations, like scattered light gradients or vignetting. Small scale changes, like near bright stars or extended galaxy halos, will be missed.

```python
#BASIC STATISTICS
#1.
meanB0, medianB0, stdB0 = sigma_clipped_stats(dataB, sigma=5.0)
print(np.array((meanB0, medianB0, stdB0)))

#-------------------------------------------------------------------------
#2.
sigma_clip = SigmaClip(sigma=5.0, maxiters=20) #creates sigma clipping object
thresholdB = detect_threshold(dataB, nsigma=2.0, sigma_clip=sigma_clip) #creates detection threshold 2 sigma above background using sigma clipping
segment_imgB = detect_sources(dataB, thresholdB, npixels=20) #creates segmentation map for all detected sources, flux above threshold.

footprint = circular_footprint(radius=10) # next 2 lines make sure galaxies dont contaminate background estimate.
mask = segment_imgB.make_source_mask(footprint=footprint)

meanB1, medianB1, stdB1 = sigma_clipped_stats(dataB, sigma=5.0, mask=mask)
print(np.array((meanB1, medianB1, stdB1)))
```

At point 1, I perform a mean median and standard deviaton of the image using sigma clipping, first without masking.

At point 2, I do the same thing but using the mask i have created.

I do this for both filters.

```python
#CREATING BACKGROUND FOR B-FILTER


plt.imshow(dataB, origin='lower', cmap='Greys_r', interpolation='nearest', vmin = 500, vmax = 700)
#interpolation = controls how pixels are drawn on the image. 
# nearest = each pixel is shown as a solid square and there is no smoothing or blending, preserves exact pixel values
plt.colorbar(label='ADU') 
plt.show()
#-----------------------------------------------------------------------------------------------

bkg_estimator = MedianBackground() #takes a median of the background values pixels
bkgB = Background2D(dataB, (1024, 1024), filter_size=(3, 3), sigma_clip=sigma_clip, bkg_estimator=bkg_estimator)
#in the median of the background it works in box sizes of 1024, splits up the image into box size of 1024
#filter size = smoothing filter of 3 x 3, smooths the background to make it look more realistic
#sigma clip = removes outliers, really bright or really dim, these outliers can bias the background estimate.
#bkg estimator = defines how background estimate of each box is computed, we chose the median estimate. 

plt.imshow(bkgB.background, origin='lower', cmap='Greys_r',interpolation='nearest')#, vmin = 5, vmax = 10 )
plt.colorbar(label='Pixel value') 
#plt.savefig('Background2D3.PDF')
plt.show()

print(f"The Background Median for the B-Filter is {bkgB.background_median:.3f}")
print(f"The Background RMS for the B-Filter is {bkgB.background_rms_median:.3f}")
```

```python
#CREATING A SEGMENTATION MAP OF ALL THE SOURCES GREATER THAN 2 SIGMA ABOVE 2D-BACKGROUND FOR B-FILTER


thresholdB = 6.5 * bkgB.background_rms
#The standard deviation of the background sky (from our earlier background estimation) = bkgB.background_rms
#Multiplying by 6.5 sets a detection threshold: only pixels that are 6.5 sigma above the background will be considered as part of sources.
#This is a high threshold, used to detect only bright sources.


kernel = make_2dgaussian_kernel(3.0, size=3)  # FWHM = 3.0
#Creates a 2D Gaussian kernel for convolution. Used to smooth the image slightly to make source detection more robust.


convolved_dataB = convolve(dataB, kernel)
#kernal = 2D array which is usually a small smoothing function
#convolve = slides the kernal across the images performing a werighted sum at each of the pixels 
#PURPOSE = enhances features of certain objects and reduces high-frequency noise, peak stars become more pronounced making source detection easier.

segment_mapB = detect_sources(convolved_dataB, thresholdB, npixels=20)
#used to detect the sources from the smoothed image created above.
#npixels = minimum number of connected pixels to be considered a source, rejects small noise peaks.

#-----------------------------------------------------------------------------------------

norm = ImageNormalize(stretch=SqrtStretch())
#ImageNormalize = Controls how pixel values map to colors when displaying an image with imshow.
#SqrtStretch = enhances faint features while keeping bright areas visible,the square root compresses high values and stretches low values


plt.imshow(dataB, origin='lower', cmap='Greys_r', norm=norm )
plt.title('Background-subtracted Data')

#-----------------------------------------------------------------------------------------

plt.imshow(segment_mapB, origin='lower', cmap=segment_mapB.cmap, interpolation='nearest')
plt.title('Segmentation Image')
plt.xlabel('Pixel Values')
plt.ylabel('Pixel Values')
plt.savefig('segmentationb.pdf')
plt.show()
```

```python
#SOURCE DEBLENDING FOR B-FILTER

segm_deblendB = deblend_sources(dataB, segment_mapB, npixels=20, nlevels=32, contrast=0.001, progress_bar=True)
#take a segmentation map where multiple sources may overlap and split them into separate objects.

plt.imshow(segm_deblendB ,  vmin = 0, vmax = 1) 
plt.gca().invert_yaxis()
plt.title('Deblended Image')
plt.xlabel('Pixel Values')
plt.ylabel('Pixel Values')
plt.savefig('deblendedb(1).pdf')
plt.show()
```

```python
#SUBTRACTING 2D IMAGE FROM SCIENCE IMAGE FOR B-FILTER

newscienceB = dataB - bkgB.background
```

```python
#CREATING SOURCE CATALOGUE FOR B-FILTER

bkgB_error = bkgB.background_rms
# each pixel contains root mean square variation of the background in its local box

errorB = photutils.utils.calc_total_error(newscienceB, bkgB_error, 2.2)
#computes total uncertainty noise and takes into account poisson noise.

catalogB = photutils.segmentation.SourceCatalog(newscienceB, segment_mapB, convolved_data=convolved_dataB, error= errorB, mask=None, background=bkgB.background, wcs= None, localbkg_width=0, apermask_method='correct', kron_params=(2.5, 1.4, 0.0), detection_cat=None, progress_bar=True)
#takes all detected sources and computes all photometric and morphological properties and takes into account background, errors, convolution for centroid refinement
# and kron apertures for flux measuremnt

tblB = catalogB.to_table(columns=('xcentroid', 'ycentroid','area', 'segment_flux', 'segment_fluxerr', 'area', 'ellipticity', 'orientation','semimajor_sigma'))
#creates a table with the columns name in the brackets, necessary for data visualisation.

print(tblB)
print(tblB[['xcentroid', 'ycentroid', 'segment_flux', 'segment_fluxerr', 'ellipticity', 'semimajor_sigma']][:10])
```

COMPUTING ZERO-POINT

To compute the zero point, we must choose a bright point on the image of Abell 2319 in the Simbad data base to then find it's coordinates. Then we could search up the name of the bright point on the internet and find its apparent magnitude. Then we use its coordinates to find it's corresponding positon values in our source table to then find its instrumental magnitude. Then to find the zero point you must subtract the instrumental magnitude from the apparent magnitude. Then we apply that zero point to all the instrumental magnitudes and then in each table to then fijnd all their apparent magnitudes. We can then compute everything from there.

Link to SIMBAD (where we found the bright point and star):
- https://simbad.u-strasbg.fr/simbad/sim-basicIdent=m33&submit=SIMBAD+search

Link to Astrometry.net (where we found the position and RA and DEC of bright point):
- https://nova.astrometry.net/upload

We had to make sure our image data didn't have a lot of noise.

```python
# NOW COMPUTING ZERO POINT AND THEN APPARENT MAGNITUDE FOR B - FILTER

m_inst = 15.4 
# reference galaxy found using above method

xcentroidb = np.asarray(tblB['xcentroid']) 
ycentroidb = np.asarray(tblB['ycentroid'])
#converting all position values into an array for easier calculation

mid_xb = np.median(xcentroidb)
mid_yb = np.median(ycentroidb)
#finding middle point of all positions/reference position

xb = tblB['m_instr'] - m_inst 

distb = np.sqrt((xcentroidb - mid_xb)**2 + (ycentroidb - mid_yb)**2)
#array of euclidian distances from each source to reference point

idxb = np.argmin(distb)
#returns index of smallest value in distb

starb = tblB[idxb]
#single row corresponding to the closest source and we can now access its properties

m_instrb = starb['m_instr']
#refernces galaxy's instrumental magnitude

print("Instrumental magnitude =", m_instrb)

ZPb = m_inst - m_instrb
#zero point calculation

print("Zero point for B is ", ZPb)

print("Closest detected source to HD7615:")
print(f"xcentroid = {starb['xcentroid']:.3f}")
print(f"ycentroid = {starb['ycentroid']:.3f}")

tblB['app_mag'] = tblB['m_instr'] + ZPb
#apparent magntiude calulation of all sources
```

```python
# NOW COMPUTING ZERO POINT AND THEN APPARENT MAGNITUDE FOR R - FILTER

m_inst = 15.4
# reference galaxy found using above method

xcentroidr = np.asarray(tblR['xcentroid'])
ycentroidr = np.asarray(tblR['ycentroid'])
#converting all position values into an array for easier calculation

mid_xr = np.median(xcentroidr)
mid_yr = np.median(ycentroidr)
#finding middle point of all positions/reference position

xr = tblR['m_instr'] - m_inst


distr = np.sqrt((xcentroidr - mid_xr)**2 + (ycentroidr - mid_yr)**2)
#array of euclidian distances from each source to reference point

idxr = np.argmin(distr)
#returns index of smallest value in distb

starr = tblR[idxr]
#single row corresponding to the closest source and we can now access its properties

m_instrr = starr['m_instr']
#refernces galaxy's instrumental magnitude

print("Instrumental magnitude =", m_instrr)

ZPr = m_inst - m_instrr
#zero point calculation
print("Zero point for R is ", ZPr)

print("Closest detected source to HD7615:")
#print(f"Label: {starb['label']}")
print(f"xcentroid = {starr['xcentroid']:.3f}")
print(f"ycentroid = {starr['ycentroid']:.3f}")

tblR['app_mag'] = tblR['m_instr'] + ZPr
#apparent magntiude calulation of all sources
```

COMPUTING COLOUR ERRORS

```python
# COMPUTING MAGNITUDE ERRORS FIRST IN ORDER TO FIND COLOUR ERRORS

tblB['mag_err'] =  (1.0857 * (tblB['segment_fluxerr'] / tblB['segment_flux'])) #error propagation
tblR['mag_err'] =  (1.0857 * (tblR['segment_fluxerr'] / tblR['segment_flux'])) 

#---------------------------------------------------------------------------------------------------

error_color = np.sqrt(tblB['mag_err'][:len(magB)]**2 + tblR['mag_err'][:len(magR)]**2)
error_R = np.asarray(tblR['mag_err'])[:len(magR)]

sigmadb = 0.1 * distance_parsec_B
tblB['abs_mag_err'] = np.sqrt(tblB['mag_err']**2 + (5/(np.log(10)) * (sigmadb / distance_parsec_B))**2) #error propagation

sigmadr = 0.1 * distance_parsec_R
tblR['abs_mag_err'] = np.sqrt(tblR['mag_err']**2 + (5/(np.log(10)) * (sigmadr / distance_parsec_R))**2)
```

What we notice is that as we approach the bluer values of the colour index , we see the error bars get larger, especially in the R-magnitude axis:

- This could be due to systematic issues, with our segmentation and deblended images meaning that the instrumental magnitude is not correct which then carries through to our errors. If the deblending doesn't split the galaxy correclty, merges two faint galaxies, misses faint flux or reads noise as real flux, then the measured instrumental flux can be wrong and becomes artificially large. This effect is emphasised in the blue end because sources become clumpy, more compact and fainter, making it difficult for the program to segment and deblend the sources cleanly. So this could inflate our errors.

- There could be that some light travels through dust, which relates to ellipticity. (Ellipticity measures how circular a galaxy is, if the ellipticity is large then the galaxy has an oval shape which means we are not viewing head-on. This could affect the magnitude as some light might be read incorrectly, some may have travelled through dust, reducing the SNR) This directly leads to the uncertainty in the colour-index. If theres uncertainty in the magnitude reading, it's almost certain that there will be uncertainty in the colour index, as the colour depends on the magnitude reading. If light travels through dust, it can appear more red on the colour index. (Dust Attenuation - dimming of light from galaxies when the light travle sthrough dust.)

CORRECTING FOR DUST EXTINCTION IN B - FILTER FIRST

```python
#APPARENT AND ABSOLUTE MAGNITUDE

Ab = 0.386                                 
m_instnewb = m_inst - Ab

tblB['corrected_mag'] = tblB['m_instr'] - Ab

xb = tblB['corrected_mag'] - m_instnewb

distb = np.sqrt((xcentroidb - mid_xb)**2 + (ycentroidb - mid_yb)**2)
idxb = np.argmin(distb)
starb = tblB[idxb]

m_instrbnew = starb['corrected_mag']
print("Instrumental magnitude =", m_instrbnew)

ZPbnew = m_instnewb - m_instrbnew
print("Zero point for B is ", ZPbnew)

print("Closest detected source to HD7615:")
#print(f"Label: {starb['label']}")
print(f"xcentroid = {starb['xcentroid']:.3f}")
print(f"ycentroid = {starb['ycentroid']:.3f}")

tblB['new_app_mag'] = tblB['corrected_mag'] + ZPbnew

#------------------------------------------------------------------------------

tblB['new_abs_mag'] = tblB['new_app_mag'] - 5*np.log10(distance_parsec_B/10)
```
